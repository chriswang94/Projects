```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import json
import requests
```

# CPI Scraper
Utilizes Beautiful Soup to extract Consumer Price Index data table from a webpage

---

```python
'''
Scrape CPI data from https://www.usinflationcalculator.com/inflation/consumer-price-index-and-annual-percent-changes-from-1913-to-2008/
'''
url = 'https://www.usinflationcalculator.com/inflation/consumer-price-index-and-annual-percent-changes-from-1913-to-2008/'
website_url = requests.get(url).text

soup = BeautifulSoup(website_url,'lxml')
```

```python
# construct dataframe out of data
cpi_df = pd.DataFrame()

col_headers = []
row=0
for tr in soup.find_all('tr'):
    if row != 0: # skip first row of table
        col = 0
        for td in tr.find_all('td'):
            if row == 1:
                col_headers.append(td.text)
                cpi_df[td.text] = np.nan # construct column headers
            else: 
                cpi_df.loc[row,col_headers[col]] = td.text
            col+=1
    row+=1
```

```python
# restructure dataframe    
cpi_df.drop(['Avg', 'Dec-Dec', 'Avg-Avg'], axis=1, inplace=True)
cpi_df.set_index('Year')
cpi_df.head(5)
```

```python
cpi_df.to_csv('cpi_df.csv', index=False)
```
